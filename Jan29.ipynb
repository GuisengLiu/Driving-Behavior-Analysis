```python
n=df.shape[0]
eRad=6371000
```

```python
import math
import numpy as np
import sys
```

```python
print(df['Latitude'][1])
```

```python
dist=[]
speed=[]
for i in range (1,n):
    var = math.sin((df['Latitude'][i]-df['Latitude'][i-1])/2)**2+math.cos(df['Latitude'][i])*math.cos(df['Latitude'][i-1])*math.sin((df['Latitude'][i]-df['Latitude'][i-1])/2)**2
    dtime = (df['Timestamp'][i]-df['Timestamp'][i-1])
    if(dtime==0):
        continue
    dist.append(np.float64(2*eRad*math.atan2(math.sqrt(var),math.sqrt(1-var))))
    speed.append(dist[-1]/dtime)
```

```python
angle=[]
for i in range (1,n):
    deltaL=df['Longitude'][i]-df['Longitude'][i-1]
    phi2=df['Latitude'][i]
    phi1=df['Latitude'][i-1]
    angle.append(math.atan2(math.sin(deltaL)*math.cos(phi2),math.cos(phi1)*math.sin(phi2)-math.sin(phi1)*math.cos(phi2)*math.cos(deltaL)))
```

```python
print(speed)
```
